# Team Work agreement

This document is meant to help you explore the dataset and set the team's ground rules for working together. Early communication and clear agreements are important to prevent conflict and navigate complicated situations smoothly. 

## Discussion prompts

Discuss the following aspects with your entire team. 

1. Learning goals and background:
    * What is your background? 
    * What is one thing you personally want to learn from this project? 
    * What is one thing you personally feel you are good at, that can contribute to the team's work? 
2. Communication norms:
    * Where will we communicate? (Whatsapp, Discord, Teams, Slack, etc)
    * When someone is stuck on the project, what should they do? How long should they struggle before asking for help?
3. Roles and rotation:
    * Who feels more comfortable with: coding, explaining results, organizing tasks, writing explanations, business aspects?
    * How will we rotate roles and divide work so that everyone learns all parts of this project? 
4. Availability & rhythm:
    * When are you realistically available to work on this project each week? 
    * What times are off-limits for you?
    * How fast do you usually reply to messages? 
5. Accountability & conflict management
    * How will we track what has been done and what's next? 
    * How will we address the case where someone is falling behind? What do you think causes team members to fall behind, and how can we avoid that and intervene early? 

## Data exploration

Consolidate your findings around the sample dataset, and decide on the next steps. 

### 1. Insights you've found 
* Which columns are relevant and which aren't? 
* What kind of business insights do you think would be worth exploring with this dataset? 
* Given the context of the challenge, what defines a "top product" for each location? 
* What columns can we create based on previous columns? 

### 2. Next steps
* Hypothesis exploration: 
    * What do you think influences sales at Panem in different locations? Do we have data around that, or could we obtain data around that?
    * Do we expect all locations to have similar sales behavior? Why or why not? 
    * Form more hypotheses you think worth exploring
* Questions:
    * Document your current unsolved/unanswered questions. If the training partner were here, what would you ask him? 
* Data manipulation:
    * Given the data, what are the next steps we should perform to clean/prepare/manipulate the dataset? 
    * What kinds of missing data could we find in the dataset? How will we deal with it?
